```python
import pandas as pd

top300 = pd.read_csv('..\\data\\movie_data\\movie_resize_ranking.csv', encoding='utf-8')
top300 = top300[['영화명','개봉일']]
```

```python
year_df = pd.DataFrame()
for row in top300['개봉일']:
    year_df = pd.concat([year_df, pd.DataFrame([row.split('-')[0]])])

year_df.columns = ['연도']

year_df.index = range(len(top300))

top300 = pd.concat([top300, year_df], axis=1).drop('')
top300
```

```python
top300[top300.duplicated()==True]
```

```python
code = pd.read_csv('..\\data\\movie_data\\concat.csv', encoding='utf-8')
code = code[['영화명','네이버URL','연도']]
```

```python
url_code = []
for i in code['네이버URL']:
    url_code.append(i.split('=')[-1])
print(url_code)
```

```python
code2 =pd.DataFrame([url_code]).transpose()
code2.columns = ['url_code']
code2
```

```python
code3 = pd.concat([code, code2],axis=1)
code3.drop_duplicates(['영화명'], keep='first', inplace=True)
code3.info()
```

```python
top300.info()
```

```python
merged_code = pd.merge(top300, code3[['영화명','url_code']], on='영화명')
merged_code.info()
```

```python
merged_code.to_csv('movie_code_utf8.csv', encoding='utf-8')
```
