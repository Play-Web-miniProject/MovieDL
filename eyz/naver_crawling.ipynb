```python
# 크롤링했던 파일 불러오기
save_file = pd.read_csv(f'save_file.csv', encoding='utf-8', index_col=0)
save_file
```

```python
# 초기화
idx = []
url_li = []
name_li = []
try:
    i = len(save_file['index'])
except:
    i = 0
print(i)
```

```python

for movie_name in movie_list[i:i+10]:
    idx.append(i)
    i+=1
    url = 'https://search.naver.com/search.naver?query=영화'+movie_name+'정보'
    res = rq.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    movie_tit = soup.find('span', class_='area_text_title')
    try:
        movie_url = movie_tit.find('a').get('href')
        url_li.append(movie_url)
        name_li.append(movie_name)
    except:
        name_li.append(movie_name)
        url_li.append('정보 없음')
    time.sleep(0.5)

print(f'{i}까지 크롤링 완료')
```

```python
tmp_df = pd.DataFrame([idx, name_li, url_li]).transpose()
tmp_df.columns = ['index', '영화명', '네이버영화 URL']
tmp_df.set_index('index')
save_file = pd.concat([save_file, tmp_df])
save_file.to_csv(f'save_file.csv', encoding='utf-8')
display(save_file)
```

```python
year_li = []
# genre_li = []
# nation_li = []
length_li = []
# open_li = []
# director_li = []
actors_li = []
# grade_li = []
rating_li = []

for url in url_li[:10]:
    res = rq.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    # 네티즌 평점 수집
    try:
        rating = soup.find_all('div', class_='score')[2].get_text()[-6:-2]
        rating_li.append(rating)
    except:
        rating_li.append('NaN')

    # 연도 수집
    year = soup.find('strong', class_='h_movie2').get_text()
    year_li.append(year.split(',')[-1].strip())
    
    info = soup.find('dl', class_='info_spec')
    dd_li = info.find_all('dd')
    dd1_span_li = dd_li[0].find_all('span')

    # 장르 수집
    # genres_li = [i.get_text() for i in dd1_span_li[0].find_all('a')]
    # genre_li.append(genres_li)

    # 국가 수집
    # nation = dd1_span_li[1].find('a')
    # nation_li.append(nation.get_text())    

    # 개봉일 수집
    # openD = dd1_span_li[3].find_all('a')
    # open_year = openD[0].get_text().strip()
    # open_day =  openD[1].get_text()
    # open_li.append(open_year+open_day)

    # 감독 수집
    # director = dd_li[1].find('a').get_text()
    # director_li.append(director)

    # 출연자 수집
    if '출연' not in [i.get_text() for i in info.find_all('dt')]:
        dd_li.insert(2,'Null')
        actors_li.append('Null')
    else:
        actors = dd_li[2].find_all('a')
        actor_li = [i.get_text() for i in actors]
        if actor_li[-1] == '더보기':
            actor_li.pop()

        actors_li.append(actor_li)

    # 상영 등급 수집
    # grade = dd_li[3].find('a').get_text()
    # grade_li.append(grade)

    # 영화 길이 수집
    length = dd1_span_li[2].get_text()
    length_li.append(length)

    time.sleep(0.5)


movie_df = pd.DataFrame([name_li, year_li, length_li, actors_li, rating_li]).transpose()
movie_df.columns = ['연도', '상영시간', '출연', '평점']

display(movie_df)
movie_df.to_csv('naver_movie.csv', encoding='utf-8')
```
