```python
# 영화명 inner join
# join200 = pd.merge(top200, df[['영화명','감독']],on='영화명')
# => 19개 데이터 손실????!!!!
```

```python
# 감독을 index 위치에 맞춰서 감독 리스트 만들기

dir_li = [0]*200
for i in range(len(top200)):
  if top200['영화명'][i] in df['영화명'].to_list():                 # 1. top200의 영화명 순서대로 하나씩 빼서(top200['영화명'][i]) df 영화명에 있는지
    index = top200['영화명'].to_list().index(top200['영화명'][i])   # 2. 있으면, 그 영화명(top200['영화명'][i])의 index 위치를 index 변수에 담기
    dir_li[index] = df['감독'][index]                               # 3. dir_li의 index 위치에 감독명을 담아주기
print(dir_li)                                                       # 결과적으로, df에서 영화명을 못찾은 영화의 감독은 0으로 남는다.
```

```python
# 위의 리스트를 DataFrame으로 만들어주고

dir_df = pd.DataFrame(dir_li)
dir_df.columns = ['감독']
dir_df
```

```python
# join200 변수에 top200과 감독 리스트를 열 기준으로 합치기
join200 = pd.concat([top200, dir_df], axis=1)
join200
```

```python
# join200 중에서 감독이 0으로 남은 것만 확인하기
join200.loc[join200['감독']==0] # df(resize 데이터)에서 없는 것
```

```python
join200.loc[join200['감독']==0].to_csv('no_data_utf-8.csv', encoding='utf-8') # csv 파일로 저장해주자.
```
